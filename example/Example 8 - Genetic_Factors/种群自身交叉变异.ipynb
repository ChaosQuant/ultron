```python
import pickle,itertools,sys,pdb
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import graphviz
from ultron.factor.genetic.genetic_factors.program import Program
from ultron.factor.genetic.genetic_factors.operators import operators_sets
from ultron.utilities.utils import check_random_state
```

```python
with open('factor_data.pkl','rb') as file2:
    total_data = pickle.load(file2)
```

```python
factor_sets = [i for i in list(set(total_data.columns)) if i not in ['trade_date','code','ret']]
random_state = check_random_state(None)
```

```python
program = Program(init_depth=3, method='full', random_state=random_state, factor_sets=factor_sets,
                  p_point=0.6,
                 n_features=2)
graphviz.Source(program.export_graphviz())
```

```python
#复制
program.reproduce()
```

```python
#树变异
program.subtree_mutation(random_state)
```

```python
#突变异
program.hoist_mutation(random_state)
```

```python
#点变异
program.point_mutation(random_state)
```

```python
# 交叉
donor = Program(init_depth=3, method='full', random_state=random_state, factor_sets=factor_sets,
                  p_point=0.6,
                 n_features=2)
program.crossover(donor._program, random_state)
```
