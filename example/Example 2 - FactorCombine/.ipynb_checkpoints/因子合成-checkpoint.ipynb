```python
from sqlalchemy import BigInteger, Column, DateTime, Float, Index, Integer, String, Text, Boolean, text, JSON,TIMESTAMP
from sqlalchemy.ext.declarative import declarative_base
from sqlalchemy import create_engine
from sqlalchemy import outerjoin, and_, select
import pandas as pd
import numpy as np
from PyFin.api import *
import pickle
import sys
sys.path.append('../..')
from ultron.factor.combine.combine_engine import CombineEngine
```

```python
begin_date = '2018-01-01'
end_date = '2018-12-31'
freq = '3b'
```

```python
rebalance_dates = makeSchedule(begin_date, end_date, freq, 'china.sse', BizDayConventions.Preceding)
```

```python
with open('combine.pkl','rb') as file2:
    experimental,market,risk_exposure,stk_returns = pickle.load(file2)
```

```python
factor_columns = [i for i in list(set(experimental.columns)) if i not in ['trade_date','code']]
```

### 等权合成

```python
%%time
equal_combine = CombineEngine.create_engine('equal_combine')
experimental['equal'] = equal_combine(experimental, factor_columns)
```

### 历史收益率加权法合成因子

```python
%%time
## 算术平均
hist_ret_combine = CombineEngine.create_engine('hist_ret_combine')
conb_df, hist_ret = hist_ret_combine(experimental, stk_returns, factor_columns, risk_exposure, [], 2)
experimental = experimental.merge(conb_df.rename(columns={'combine':'hist_ret_equal'}), on=['code','trade_date'])
```

```python
%%time
## 半衰加权
conb_df, hist_ret = hist_ret_combine(experimental, stk_returns, factor_columns, risk_exposure, [], 2,
                                    'half_life', 2)
experimental = experimental.merge(conb_df.rename(columns={'combine':'hist_ret_half_life'}), 
                                  on=['code','trade_date'])
```

```python
experimental
```

### 历史IC加权法合成因子

```python
%%time
## 算术平均
hist_ic_combine = CombineEngine.create_engine('hist_ic_combine')
conb_df, hist_ret = hist_ic_combine(experimental, stk_returns, factor_columns, 2)
experimental = experimental.merge(conb_df.rename(columns={'combine':'hist_ic_equal'}), 
                                  on=['code','trade_date'])
```

```python
%%time
## 半衰加权
hist_ic_combine = CombineEngine.create_engine('hist_ic_combine')
conb_df, hist_ret = hist_ic_combine(experimental, stk_returns, factor_columns, 4, 'half_life', 2)
experimental = experimental.merge(conb_df.rename(columns={'combine':'hist_ic_half_life'}), 
                                  on=['code','trade_date'])
```

### 最大化ICIR加权法合成因子

```python
%%time
##  估计协方差矩阵的方法
max_icir_combine = CombineEngine.create_engine('max_icir_combine')
conb_df, m_ir_df = max_icir_combine(experimental, stk_returns, factor_columns, 4)
experimental = experimental.merge(conb_df.rename(columns={'combine':'max_icir_sample'}), 
                                  on=['code','trade_date'])
```

```python
%%time
##  压缩估计协方差
max_icir_combine = CombineEngine.create_engine('max_icir_combine')
conb_df, m_ir_df = max_icir_combine(experimental, stk_returns, factor_columns, 4, 'shrunk', False)
experimental = experimental.merge(conb_df.rename(columns={'combine':'max_icir_shrunk_0'}), 
                                  on=['code','trade_date'])
```

```python
%%time
##  压缩估计协方差 限制权重为正
max_icir_combine = CombineEngine.create_engine('max_icir_combine')
conb_df, m_ir_df = max_icir_combine(experimental, stk_returns, factor_columns, 4, 'shrunk', True)
experimental = experimental.merge(conb_df.rename(columns={'combine':'max_icir_shrunk_1'}), 
                                  on=['code','trade_date'])
```

### 最大化IC加权法合成因子

```python
%%time
## 样本协方差
max_ic_combine = CombineEngine.create_engine('max_ic_combine')
conb_df, m_ir_df = max_ic_combine(experimental, stk_returns, factor_columns, 4, 'sample')
experimental = experimental.merge(conb_df.rename(columns={'combine':'max_ic_sample'}), 
                                  on=['code','trade_date'])
```

```python
%%time
## 压缩估计协方差 约束权重为负
max_ic_combine = CombineEngine.create_engine('max_ic_combine')
conb_df, m_ir_df = max_ic_combine(experimental, stk_returns, factor_columns, 4, 'shrunk', False)
experimental = experimental.merge(conb_df.rename(columns={'combine':'max_ic_shrunk_0'}), 
                                  on=['code','trade_date'])
```

```python
%%time
## 压缩估计协方差 约束权重为正
max_ic_combine = CombineEngine.create_engine('max_ic_combine')
conb_df, m_ir_df = max_ic_combine(experimental, stk_returns, factor_columns, 4, 'shrunk', True)
experimental = experimental.merge(conb_df.rename(columns={'combine':'max_ic_shrunk_1'}), 
                                  on=['code','trade_date'])
```
