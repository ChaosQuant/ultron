### 因子合成，用性能指标的结果为分数

```python
##  最大化ICIR加权法合成因子 压缩估计协方差 限制权重为正 合成因子。用IC作为目标值
def objective(total_data, sub_col):
    factor_data = total_data[list(sub_col) + ['code','trade_date']]
    stk_returns = total_data[['code','trade_date','ret']]
    stk_returns['nxt1_ret'] = stk_returns.groupby('code')['ret'].shift(-1)
    max_icir_combine = CombineEngine.create_engine('max_icir_combine')
    conb_df, m_ir_df = max_icir_combine(factor_data, stk_returns, list(sub_col), 4, 'shrunk', True)
    factor_data = factor_data.merge(conb_df,on=['code','trade_date']).merge(stk_returns, 
                                            on=['code','trade_date'])
    return factor_data[['combine','ret']].corr(method='spearman').values[0, 1]
```

```python
genetic = GeneticFactors(0.2, 0.4, 0.4,
                        10, 20, 20, objective) # 构造遗传算法选因子
```

```python
factor_columns = [i for i in list(set(factors_data.columns)) if i not in ['trade_date','code','ret']]
```

```python
## 可采用其他方式来进行特征分群, 目前采用随机
point = int(np.random.uniform(0, len(factor_columns))/2)
ori_field = factor_columns[:point]
add_field = factor_columns[point:]
```

```python
genetic.genetic_run(factors_data, None, ori_field, add_field)
```
